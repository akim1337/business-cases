## Что понятно из данных
**Как происходит заполнение данных:**
1. Пользователь создал коризину со временем create_date и basket_id
2. Начал оформляться, скорее всего было промежуточное update_date
3. Оформился в течении часа - получил data, id,  update_date
4. Не оформился в течении часа - получил status 5. Но, если происходило оформление заказа снова после часа, то получал update_date в новой строке с новым basket_id. Старый update_date имеет последнее обновление при получении статуса 5
5. Оформление нового заказа, начало в time_, получение нового статуса update_date в новой строке, с новой корзиной

**Артефакты:**
1. Проверка статуса заказа /?headerId=498145316  добавляет новую строку в gbq с информацией о том, когда пользователь просмотрел информацию о заказе
2. 2023-11-27 20:42 - у таких пользователей статуса нет, т.к обрезается выборка на этом числе! важно учитывать, что 27 не полный. Лефт соединение уберет у нас такие данные


**Что можем получить из данных:**
1. *Время оформления успешных заказов*: update_date - time_  одного статуса 4
2. Разница между update_date и началом нового оформления time_ (update_date статуса 5 и time_ статуса 4)
3. *Время повторного оформления при провальной первой попытке*: update_date + (update_date минус time_) + (update_date статуса 5 минус time_ статуса 4), где:
*   update_date - время присвоения 5 статуса, что значит - отведенный час истек, добавляем 60 минут
*   (update_date -  time_) время оформления нового(старого) заказа, он же успешный заказ
*   (time_ - update_date (поймем, когда человек пришел еще раз для оформления. Тайм берем 4 статуса, а апдейт - 5го)
4. *Проще про пункт 3*, update_date 5го статуса - update_date 4го статуса одного и того же заказа - время между провальной попыткой и успешной попыткой. Искомая величина для ответа на поставленную задачу, т.к. даст понимание того, как распределены оформления заказов по времени.

## Чистка данных

```python
# Переводим все даты в формат дат
gbq['time_'] = pd.to_datetime(gbq['time_'], format='%Y-%m-%d %H:%M:%S')
crm['update_date'] = pd.to_datetime(crm['update_date'], format='%Y-%m-%d %H:%M:%S')
crm['create_date'] = pd.to_datetime(crm['create_date'], format='%Y-%m-%d %H:%M:%S')
```

```python
# Дропаем все строки с ?headerId=1245498347 т.к будет передавать лишние дубли в итоговую таблицу
gbq_drop_headerId = gbq[~gbq['pagepath'].str.contains('headerId=', regex=False)]
```

```python
gbq_drop_headerId.count()
```

```python
# Заберем даты, для дополнительного совпдаения при мерже таблицы
crm['update_date_y_m_d'] = pd.to_datetime(crm['update_date']).dt.date
```

```python
crm.head()
```

```python
# CRM мержу саму с собой, чтобы найти последовательности сделок, у которых статус 5, но которые заказали то же самое, и в итоге оформились.
# В целом, нам интересны те пользователи, которые сразу сделали перезаказ, поэтому соединять будем именно по значениям 'sum_price', 'sum_points', т.к. таким образом мы нивелируем разность дат и уберем тех пользователей, которые не оформились по каким либо другим причинам. Здесь выборка будет максимально целевая.
crm_status_5to4 = crm.merge(crm, how='left', left_on=['contract', 'sum_price', 'sum_points'], right_on=['contract', 'sum_price', 'sum_points'])
crm_status_5to4 = crm_status_5to4.loc[:, ['basket_id_x', 'status_x', 'id_x', 'contract', 'sum_price', 'sum_points', 'create_date_x', 'update_date_x', 'status_y', 'basket_id_y', 'create_date_y', 'update_date_y']]
```

```python
# к CRM мержу данный с gbq, забираю только time_, т.к. остальные столбцы не понадобятся
df = crm_status_5to4.merge(gbq_drop_headerId, how='left', left_on=['contract', 'basket_id_y'], right_on=['contract', 'basket_id'])
df = df.loc[:, ['basket_id_x', 'status_x', 'id_x', 'contract', 'sum_price', 'sum_points', 'create_date_x', 'update_date_x', 'status_y', 'basket_id_y', 'create_date_y', 'update_date_y', 'time_']]
```

```python
import matplotlib.pyplot as plt
table_uniq['hours'].hist(bins=24, range=(0, 24))
plt.xlabel('Hours')
plt.ylabel('Frequency')
plt.title('Распределение провальной оплаты в сутках по часам')
plt.show()
```
